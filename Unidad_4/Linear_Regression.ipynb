```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
auto = pd.read_csv('/DiabetesClassification.csv')
auto.head()
```

```python
from sklearn.linear_model import LinearRegression

# Load the data from the auto dataframe
X = auto['BMI'].values.reshape(-1, 1)
y = auto['Age'].values

# Fit a linear regression line to the data
model = LinearRegression()
model.fit(X, y)

# Extract the coefficients
w = model.coef_[0]
b = model.intercept_

# Print the equation
print(f'Y = {w}x + {b}')
```

```python
plt.figure(figsize=(10,10))
plt.scatter(auto['BMI'],auto['Age'])
plt.title('BMI in people')
plt.xlabel('age')
plt.ylabel('BMI')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(auto['BMI'],auto['Age'])
x_values = range(50)  # Ejemplo de valores de x
y_values = [0.9106565605219344 * x + 26.535952046202524 for x in x_values]
# Graficar la ecuación
plt.plot(x_values, y_values, c='red', label='y = 0.9106565605219344x + 26.535952046202524')
plt.title('BMI in people')
plt.xlabel('age')
plt.ylabel('BMI')
plt.show()
```
