```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
auto = pd.read_csv('/content/DiabetesClassification.csv')
auto.head()
```

```python
X = auto[['Age']]
Y = auto['BMI']
```

```python
MPG_Pred = LinearRegression()
MPG_Pred.fit(X,Y)
plt.figure(figsize=(10,10))
plt.scatter(auto['Age'], auto['BMI'])
plt.scatter(X,MPG_Pred.predict(X), c='Red')
plt.title('BMI in people')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.show()
```
